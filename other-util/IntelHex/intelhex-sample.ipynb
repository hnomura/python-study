# IntelHex 

Documentation at http://python-intelhex.readthedocs.io/en/stable/index.html

This is just a simple example. 
NEW (APP & BLD) are from MSP430HEX utility with options removing duplicated sections in APP.   
OLD (APP & BLD) have duplicated memory sections between APP and BLD. 


```python
from intelhex import IntelHex

hex_app_new = IntelHex() 
hex_app_new.loadhex('./new/PROG_TAIKO_LTS_ENP.hex')

hex_bld_new = IntelHex() 
hex_bld_new.loadhex('./new/BLDR_TAIKO_LTS_ENP.hex')

hex_app_old = IntelHex() 
hex_app_old.loadhex('./old/PROG_TAIKO_LTS_ENP.hex')

hex_bld_old = IntelHex()
hex_bld_old.loadhex('./old/BLDR_TAIKO_LTS_ENP.hex')
```

```python
# Get memory segments 
seg_app_new = hex_app_new.segments()
seg_bld_new = hex_bld_new.segments()
seg_app_old = hex_app_old.segments()
seg_bld_old = hex_bld_old.segments()
```

```python
# Pring segment in Hexadecimal range 
print('APP NEW ', [('0x'+format(s,'04x'), '0x'+format(e,'04x')) for s,e  in seg_app_new])
print('BLD NEW ', [('0x'+format(s,'04x'), '0x'+format(e,'04x')) for s,e  in seg_bld_new])
print()
print('APP OLD ', [('0x'+format(s,'04x'), '0x'+format(e,'04x')) for s,e  in seg_app_old])
print('BLD OLD ', [('0x'+format(s,'04x'), '0x'+format(e,'04x')) for s,e  in seg_bld_old])
```

```python
# Dump APP NEW image, from 0x2600 to 0x2800 (512 bytes)
# Note that holes will show 255 (0xFF) 
start = int("2600",16)

for i in range(start,start+512):
    print('0x'+format(i,'04X'), ':', '0x'+format(hex_app_new[i],'04X'))
    
```
